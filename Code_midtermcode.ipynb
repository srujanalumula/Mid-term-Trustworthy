```python
# ---------------------------------------------------
# 📌 IMPORTING ESSENTIAL LIBRARIES
# ---------------------------------------------------
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from tqdm.notebook import tqdm

# TensorFlow and Keras modules for Deep Learning
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Dense, Dropout, Flatten, Conv2D, MaxPooling2D, BatchNormalization
)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import (
    ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
)
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Additional utilities
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import warnings
warnings.filterwarnings("ignore")

# Kaggle dataset download module
import kagglehub


# ---------------------------------------------------
# 📂 DOWNLOADING THE DATASET
# ---------------------------------------------------
dataset_path = kagglehub.dataset_download("masoudnickparvar/white-blood-cells-dataset")
print("✅ Dataset downloaded successfully at:", dataset_path)

# Setting random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print("\nInitializing dataset directory paths...")

# ---------------------------------------------------
# 🗂️ DEFINE TRAIN AND TEST PATHS
# ---------------------------------------------------
train_dir = os.path.join(dataset_path, "Train")
test_dir = os.path.join(dataset_path, "Test-A")
class_names = ["Neutrophil", "Lymphocyte", "Monocyte", "Eosinophil", "Basophil"]

# Optional: Print the dataset structure
for dirpath, _, filenames in os.walk('/kaggle/input'):
    print(dirpath)


# ---------------------------------------------------
# 🧩 FUNCTION TO LOAD IMAGES AND LABELS
# ---------------------------------------------------
def load_images(directory, class_list):
    """
    Loads all images from given directory.
    Converts them to RGB and resizes to 128x128.
    Returns NumPy arrays of images, labels, and paths.
    """
    images, labels, paths = [], [], []
    target_size = 128

    for label_idx, class_name in enumerate(class_list):
        class_path = os.path.join(directory, class_name)

        # Skip missing folders
        if not os.path.exists(class_path):
            print(f"⚠️ Missing path: {class_path}")
            continue

        # Loop through all image files
        for img_file in tqdm(os.listdir(class_path), desc=f"Loading {class_name}"):
            img_path = os.path.join(class_path, img_file)
            try:
                img = Image.open(img_path).convert("RGB")
                img = img.resize((target_size, target_size))
                images.append(np.array(img))
                labels.append(label_idx)
                paths.append(img_path)
            except Exception as e:
                print(f"❌ Failed to load {img_path}: {e}")

    return np.array(images), np.array(labels), paths


# ---------------------------------------------------
# 📥 LOADING TRAIN AND TEST DATA
# ---------------------------------------------------
X_train_raw, y_train_raw, train_paths = load_images(train_dir, class_names)
X_test_raw, y_test_raw, test_paths = load_images(test_dir, class_names)

print(f"\nTraining images loaded: {len(X_train_raw)}")
print(f"Testing images loaded: {len(X_test_raw)}")

# ---------------------------------------------------
# 🔍 EXPLORING DATA DISTRIBUTION
# ---------------------------------------------------
print("\nAnalyzing class distribution...")

train_counts = np.bincount(y_train_raw)
for i, count in enumerate(train_counts):
    print(f"{class_names[i]}: {count} samples")

plt.figure(figsize=(10, 5))
sns.countplot(x=y_train_raw)
plt.title("Training Data Class Distribution")
plt.xticks(range(len(class_names)), class_names, rotation=45)
plt.xlabel("Cell Type")
plt.ylabel("Count")
plt.show()


# ---------------------------------------------------
# 🖼️ DISPLAY SAMPLE IMAGES FROM EACH CLASS
# ---------------------------------------------------
plt.figure(figsize=(15, 8))
for idx, label in enumerate(class_names):
    if idx < len(class_names):
        indices = np.where(y_train_raw == idx)[0]
        if len(indices) > 0:
            plt.subplot(1, len(class_names), idx + 1)
            plt.imshow(X_train_raw[indices[0]])
            plt.title(label)
            plt.axis("off")
plt.tight_layout()
plt.show()


# ---------------------------------------------------
# 🧮 NORMALIZATION AND ONE-HOT ENCODING
# ---------------------------------------------------
print("\nPreprocessing data for model input...")

X_train = X_train_raw / 255.0
X_test = X_test_raw / 255.0

y_train = tf.keras.utils.to_categorical(y_train_raw, len(class_names))
y_test = tf.keras.utils.to_categorical(y_test_raw, len(class_names))

# Splitting validation data
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42, stratify=y_train
)

print(f"Training images: {len(X_train)}")
print(f"Validation images: {len(X_val)}")
print(f"Testing images: {len(X_test)}")


# ---------------------------------------------------
# 📈 DATA AUGMENTATION
# ---------------------------------------------------
print("\nApplying data augmentation...")

augmenter = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode="nearest"
)

# Visualizing sample augmentations
plt.figure(figsize=(15, 5))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    aug_img = augmenter.random_transform(X_train[0])
    plt.imshow(aug_img)
    plt.axis("off")
plt.suptitle("Sample Augmented Images")
plt.show()


# ---------------------------------------------------
# 🧠 BUILDING A CNN MODEL
# ---------------------------------------------------
print("\nConstructing the CNN architecture...")

def build_cnn_model():
    """
    Custom high-performance CNN model for WBC classification.
    """
    model = Sequential([
        # Block 1
        Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=(128, 128, 3)),
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.25),

        # Block 2
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.25),

        # Block 3
        Conv2D(256, (3, 3), activation='relu', padding='same'),
        Conv2D(256, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.3),

        # Block 4
        Conv2D(512, (3, 3), activation='relu', padding='same'),
        Conv2D(512, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.4),

        # Dense Layers
        Flatten(),
        Dense(1024, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(512, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),

        # Output Layer
        Dense(len(class_names), activation='softmax')
    ])

    model.compile(
        optimizer=Adam(learning_rate=0.0005),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model


cnn_model = build_cnn_model()
cnn_model.summary()


# ---------------------------------------------------
# 🎯 TRAINING THE MODEL
# ---------------------------------------------------
callbacks = [
    ModelCheckpoint('best_wbc_model.keras', save_best_only=True, monitor='val_accuracy', mode='max'),
    EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)
]

print("\nStarting training process...")
history = cnn_model.fit(
    augmenter.flow(X_train, y_train, batch_size=32),
    validation_data=(X_val, y_val),
    epochs=10,
    callbacks=callbacks,
    verbose=1
)


# ---------------------------------------------------
# 📊 EVALUATION ON TEST SET
# ---------------------------------------------------
test_loss, test_acc = cnn_model.evaluate(X_test, y_test, verbose=1)
print(f"\n✅ Test Accuracy: {test_acc:.4f}")


# ---------------------------------------------------
# 📈 METRICS AND VISUALIZATIONS
# ---------------------------------------------------
pred_probs = cnn_model.predict(X_test)
y_pred_classes = np.argmax(pred_probs, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

print("\nClassification Report:")
print(classification_report(y_true_classes, y_pred_classes, target_names=class_names))

# Confusion Matrix
plt.figure(figsize=(10, 8))
cm = confusion_matrix(y_true_classes, y_pred_classes)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()


# Accuracy/Loss Curves
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label="Train Accuracy")
plt.plot(history.history['val_accuracy'], label="Val Accuracy")
plt.legend()
plt.title("Accuracy per Epoch")

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label="Train Loss")
plt.plot(history.history['val_loss'], label="Val Loss")
plt.legend()
plt.title("Loss per Epoch")
plt.show()


# ---------------------------------------------------
# 🔮 PREDICTION ON SAMPLE IMAGE
# ---------------------------------------------------
def predict_single_image(img_path, model=cnn_model):
    """
    Predicts a single image and visualizes class probabilities.
    """
    img = Image.open(img_path).convert("RGB").resize((128, 128))
    img_arr = np.expand_dims(np.array(img) / 255.0, axis=0)
    preds = model.predict(img_arr)
    pred_idx = np.argmax(preds, axis=1)[0]

    print(f"🧫 Prediction: {class_names[pred_idx]} ({preds[0][pred_idx]*100:.2f}%)")

    # Display image
    plt.figure(figsize=(6, 6))
    plt.imshow(img)
    plt.title(f"Predicted: {class_names[pred_idx]}")
    plt.axis("off")
    plt.show()

    # Probability bar chart
    plt.figure(figsize=(10, 3))
    sns.barplot(x=class_names, y=preds[0])
    plt.title("Prediction Probabilities")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    return class_names[pred_idx], preds[0]


if len(test_paths) > 0:
    random_idx = np.random.choice(range(len(test_paths)))
    print("\nTesting model on a random test image...")
    predict_single_image(test_paths[random_idx])

print("Project completed successfully")
```

```python
# ---------------------------------------------------
# 📥 LOADING 224x224 IMAGES FOR VGG16/ResNet50
# ---------------------------------------------------
from tensorflow.keras.applications import VGG16, ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras import layers

def load_images_224(directory, class_list):
    """Load images at 224x224 resolution"""
    images, labels = [], []
    for label_idx, class_name in enumerate(class_list):
        class_path = os.path.join(directory, class_name)
        if not os.path.exists(class_path):
            continue
        for img_file in tqdm(os.listdir(class_path), desc=f"Loading {class_name}"):
            try:
                img = Image.open(os.path.join(class_path, img_file)).convert("RGB")
                img = img.resize((224, 224))
                images.append(np.array(img))
                labels.append(label_idx)
            except:
                pass
    return np.array(images), np.array(labels)

print("Loading 224x224 images...")
X_train_224, y_train_224 = load_images_224(train_dir, class_names)
X_test_224, y_test_224 = load_images_224(test_dir, class_names)

# Normalize
X_train_224 = X_train_224 / 255.0
X_test_224 = X_test_224 / 255.0

# One-hot encode
y_train_224 = tf.keras.utils.to_categorical(y_train_224, 5)
y_test_224 = tf.keras.utils.to_categorical(y_test_224, 5)

# Split
X_train_224, X_val_224, y_train_224, y_val_224 = train_test_split(
    X_train_224, y_train_224, test_size=0.2, random_state=42, stratify=y_train_224
)

# Augmentation
augmenter_224 = ImageDataGenerator(
    rotation_range=20, width_shift_range=0.2, height_shift_range=0.2,
    horizontal_flip=True, vertical_flip=True
)

print(f"✅ 224x224 Images loaded: Train={len(X_train_224)}, Test={len(X_test_224)}")
```

```python
# ---------------------------------------------------
# 🧠 VGG16 MODEL
# ---------------------------------------------------
print("\n" + "="*60)
print("TRAINING VGG16 MODEL")
print("="*60)

# Build VGG16
vgg_base = VGG16(input_shape=(224,224,3), include_top=False, weights='imagenet')
for layer in vgg_base.layers:
    layer.trainable = False

x = layers.Flatten()(vgg_base.output)
x = layers.Dense(200, activation='relu')(x)
x = layers.Dropout(0.5)(x)
output = layers.Dense(5, activation='softmax')(x)
vgg_model = Model(vgg_base.input, output)

vgg_model.compile(
    optimizer=Adam(0.0001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\nVGG16 Architecture:")
vgg_model.summary()

# Train
history_vgg = vgg_model.fit(
    augmenter_224.flow(X_train_224, y_train_224, batch_size=32),
    validation_data=(X_val_224, y_val_224),
    epochs=10,
    callbacks=[EarlyStopping(patience=5, restore_best_weights=True)],
    verbose=1
)

vgg_model.save('vgg16_wbc_model.keras')
print("\n✅ VGG16 model saved!")
```

```python
# ---------------------------------------------------
# 📊 VGG16 EVALUATION
# ---------------------------------------------------
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("\n" + "="*60)
print("VGG16 RESULTS")
print("="*60)

loss_vgg, acc_vgg = vgg_model.evaluate(X_test_224, y_test_224, verbose=0)
y_pred_vgg = np.argmax(vgg_model.predict(X_test_224, verbose=0), axis=1)
y_true_vgg = np.argmax(y_test_224, axis=1)

print(f"\nAccuracy:  {acc_vgg:.4f}")
print(f"Precision: {precision_score(y_true_vgg, y_pred_vgg, average='weighted'):.4f}")
print(f"Recall:    {recall_score(y_true_vgg, y_pred_vgg, average='weighted'):.4f}")
print(f"F1-Score:  {f1_score(y_true_vgg, y_pred_vgg, average='weighted'):.4f}")

print("\nClassification Report:")
print(classification_report(y_true_vgg, y_pred_vgg, target_names=class_names))

# Confusion Matrix
plt.figure(figsize=(8, 6))
cm_vgg = confusion_matrix(y_true_vgg, y_pred_vgg)
sns.heatmap(cm_vgg, annot=True, fmt='d', cmap='Greens',
            xticklabels=class_names, yticklabels=class_names)
plt.title(f'VGG16 Confusion Matrix (Acc: {acc_vgg:.3f})')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()

# Training curves
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history_vgg.history['accuracy'], label='Train')
plt.plot(history_vgg.history['val_accuracy'], label='Validation')
plt.title('VGG16 Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history_vgg.history['loss'], label='Train')
plt.plot(history_vgg.history['val_loss'], label='Validation')
plt.title('VGG16 Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ---------------------------------------------------
# 🧠 RESNET50 MODEL
# ---------------------------------------------------
print("\n" + "="*60)
print("TRAINING RESNET50 MODEL")
print("="*60)

# Build ResNet50
resnet_base = ResNet50(input_shape=(224,224,3), include_top=False, weights='imagenet')
for layer in resnet_base.layers:
    layer.trainable = False

z = layers.Flatten()(resnet_base.output)
z = layers.Dense(150, activation='relu')(z)
output = layers.Dense(5, activation='softmax')(z)
resnet_model = Model(resnet_base.input, output)

resnet_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\nResNet50 Architecture:")
resnet_model.summary()

# Train
history_resnet = resnet_model.fit(
    augmenter_224.flow(X_train_224, y_train_224, batch_size=64),
    validation_data=(X_val_224, y_val_224),
    epochs=10,
    callbacks=[EarlyStopping(patience=5, restore_best_weights=True)],
    verbose=1
)

resnet_model.save('resnet50_wbc_model.keras')
print("\n✅ ResNet50 model saved!")
```

```python
# ---------------------------------------------------
# MODEL COMPARISON
# ---------------------------------------------------
print("\n" + "="*60)
print("FINAL COMPARISON - ALL MODELS")
print("="*60)

# Get CNN results (already evaluated in your code)
results = pd.DataFrame({
    'Model': ['Custom CNN', 'VGG16', 'ResNet50'],
    'Accuracy': [test_acc, acc_vgg, acc_resnet],
    'Loss': [test_loss, loss_vgg, loss_resnet],
    'Input Size': ['128x128', '224x224', '224x224']
})

print("\n")
print(results.to_string(index=False))

# Bar chart comparison
plt.figure(figsize=(10, 6))
plt.bar(results['Model'], results['Accuracy'], color=['#3498db', '#2ecc71', '#e74c3c'])
plt.title('Model Accuracy Comparison', fontsize=16, fontweight='bold')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
for i, row in results.iterrows():
    plt.text(i, row['Accuracy'] + 0.02, f"{row['Accuracy']:.4f}",
             ha='center', fontsize=12, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("\n🎉 PROJECT COMPLETED - ALL MODELS TRAINED AND EVALUATED!")
print("\n📦 Saved Models:")
print("   1. best_wbc_model.keras (Custom CNN)")
print("   2. vgg16_wbc_model.keras")
print("   3. resnet50_wbc_model.keras")
```
